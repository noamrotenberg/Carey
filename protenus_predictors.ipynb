```python
df['Quarter'] = df['Snapshot_Date'].dt.to_period('Q')
print(len(df['Opportunity_ID'].unique()))
print(len(df['Account_Zip_Code'].unique()))
print(len(df['Account_Cash_on_Hand'].unique()))
import sklearn.model_selection
import copy
import matplotlib.pyplot as plt

unique_OPIDs = df['Opportunity_ID'].unique()
df['Opportunity_Created_Quarter'] = df['Opportunity_Created_Date'].dt.to_period('Q')

OPIDs_cannot_be_split = list()
for j in [df[df['Opportunity_ID'] == OPID].iloc[0,:]['Opportunity_Created_Date'].to_period('Q') for OPID in unique_OPIDs]:
  if 1 == (len(df[j == df['Opportunity_Created_Date'].dt.to_period('Q')]['Opportunity_ID'].unique())):
    print(j, df[j == df['Opportunity_Created_Date'].dt.to_period('Q')]['Opportunity_ID'].unique())
    OPIDs_cannot_be_split.append(df[j == df['Opportunity_Created_Date'].dt.to_period('Q')]['Opportunity_ID'].unique()[0])

unique_OPIDs_to_split = list(copy.deepcopy(unique_OPIDs))
for OPID in OPIDs_cannot_be_split: unique_OPIDs_to_split.remove(OPID)

train_OPIDs, test_OPIDs = sklearn.model_selection.train_test_split(unique_OPIDs_to_split, test_size=0.2, stratify=[df[df['Opportunity_ID'] == OPID].iloc[0,:]['Opportunity_Created_Quarter'] for OPID in unique_OPIDs_to_split], random_state=0)
train_OPIDs, val_OPIDs  = sklearn.model_selection.train_test_split(train_OPIDs, train_size=0.6/0.8, stratify=[df[df['Opportunity_ID'] == OPID].iloc[0,:]['Opportunity_Created_Quarter'] for OPID in train_OPIDs], random_state=0)

# assign OPIDs_cannot_be_split:
train_OPIDs.append(OPIDs_cannot_be_split[0])
val_OPIDs.append(OPIDs_cannot_be_split[1])

# make a df that has only unique OPIDs:
indices_of_1st_snaps = [df[df['Opportunity_ID']==OPID]['Snapshot_Date'].sort_values().index[0] for OPID in unique_OPIDs]
df_1st_snaps = df.loc[indices_of_1st_snaps]
df_last_snaps = df.loc[[df[df['Opportunity_ID']==OPID]['Snapshot_Date'].sort_values().index[-1] for OPID in unique_OPIDs]]

plt.title("Distribution of quarters across sets")
for OPIDs in [train_OPIDs, val_OPIDs, test_OPIDs]:
  plt.hist(df_1st_snaps[df_1st_snaps['Opportunity_ID'].isin(OPIDs)]['Opportunity_Created_Quarter'].astype(str), alpha=0.5, bins=len(df['Opportunity_Created_Quarter'].unique()), range=(0,len(df['Opportunity_Created_Quarter'].unique())))
plt.xticks(rotation=90)
plt.ylabel("Number of new OPIDs created")
plt.xlabel("Quarter")
plt.show()

```

```python
import numpy as np
df_X = pd.DataFrame()
df_y = pd.Series()

def get_y(row, df_q2):
  closed_won  = np.sum(df_q2[df_q2['Opportunity_ID']==row['Opportunity_ID']]['Opportunity_Stage'] == 'Closed Won') > 0
  closed_lost = np.sum(df_q2[df_q2['Opportunity_ID']==row['Opportunity_ID']]['Opportunity_Stage'] == 'Closed Lost') > 0

  if closed_won:
    return 1
  elif closed_lost:
    return -1
  else:
    return 0

for quarter1, quarter2 in zip(df['Quarter'].unique()[:-1], df['Quarter'].unique()[1:]): # maybe change to month0 & months3
  df_q2 = df[(df['Quarter'] == quarter2)]
  df_q2_unique_OPIDs = df_q2['Opportunity_ID'].unique()
  df_q1 = df[(df['Quarter'] == quarter1) & ~(df['Opportunity_Stage'].isin(['Closed Won', 'Closed Lost'])) & (df['Opportunity_ID'].isin(df_q2_unique_OPIDs))]
  df_X = df_X.append(df_q1)

  df_y = df_y.append(df_q1.apply(lambda row: get_y(row, df_q2), axis=1))

for quarter1 in df['Quarter'].unique(): # maybe change to month0 & months3
  df_q2 = df[(df['Quarter'] == quarter2)]
  df_q2_unique_OPIDs = df_q2['Opportunity_ID'].unique()
  df_q1 = df[(df['Quarter'] == quarter1) & ~(df['Opportunity_Stage'].isin(['Closed Won', 'Closed Lost'])) & (df['Opportunity_ID'].isin(df_q2_unique_OPIDs))]
  df_X = df_X.append(df_q1)



# make data numerical
df_X['Opportunity_Stage_n'] = df_X['Opportunity_Stage'].replace(['Stage 0 - Prospecting', 'Stage 1 - Evaluating', 'Stage 2 - Scoping', 'Stage 3 - Pricing', 'Stage 4 - Verbal / VOC', 'Stage 5 - Contracting'],
                                                                  [0, 1, 2, 3, 4, 5])
df_X['Opportunity_Product_n'] = df_X['Opportunity_Product'].replace(['Diversion Monitoring', 'Privacy Monitoring', 'Compliance Analytics'],
                                                                    [1, 2, 3])
# what does Opportunity_Product 0 mean??

df_X['Account_Electronic_Medical_Record_System_n'] = df_X['Account_Electronic_Medical_Record_System'].replace(['Epic', 'Cerner', 'Meditech', 'Other'],
                                                                                                              [1, 2, 3, -1])

df_X['Account_Hospital_Type_n'] = df_X['Account_Hospital_Type'].replace(['Childrens Hospital', 'Health System', 'Short Term Acute Care Hospital', 'Critical Access Hospital',
       'Psychiatric Hospital', 'Rehabilitation Hospital', 'Long Term Acute Care Hospital'],
                                                                        [0, 1, 2, 3] + [-1, -1, -1])
df_X['Account_Hospital_Type_n'].fillna(0, inplace=True)
df_X['Opportunity_Annual_Amount'].fillna(0, inplace=True)
df_X[['Account_Number_of_Hospitals ', 'Account_Number_of_Beds', 'Account_Number_of_Affiliated_Physicians', 'Account_Number_of_Employee',
     'Account_Cash_on_Hand', 'Account_Total_Patient_Revenue', 'Account_Net_Patient_Revenue']].fillna(0, inplace=True)
df_X.fillna(0, inplace=True)

df_X['Opportunity_Type_n'] = df['Opportunity_Type'].replace(['New Business', 'Existing Business'],
                                                            [0, 1])

df_X['Account_Hospital_Type'].describe()

df_X.columns
for col in df_X.columns:
  print(col, df_X[col].isnull().values.any())

# additional features to use:
# Account_State geographic region
# Account_Zip_Code socioeconomic status
# Account_Fiscal_Year_End to float

df_X['time_since_created_date'] = (df_X['Snapshot_Date'] - df_X['Opportunity_Created_Date']).astype(int)
df_X['time_until_est_close_date'] = (df_X['Opportunity_Close_Date'] - df_X['Snapshot_Date']).astype(int)

df_X_trimmed = df_X[['Opportunity_Annual_Amount',
       'Account_Number_of_Hospitals ',
       'Account_Number_of_Beds', 'Account_Number_of_Affiliated_Physicians',
       'Account_Number_of_Employee', 'Account_Cash_on_Hand',
       'Account_Total_Patient_Revenue', 'Account_Net_Patient_Revenue',
       'Opportunity_Stage_n',
       'Opportunity_Product_n', 'Account_Electronic_Medical_Record_System_n',
       'Account_Hospital_Type_n', 'Opportunity_Type_n', 'time_since_created_date',
                     'time_until_est_close_date']]

df_trainX = df_X_trimmed[df_X['Opportunity_ID'].isin(train_OPIDs)]
df_valX   = df_X_trimmed[df_X['Opportunity_ID'].isin(val_OPIDs)]
df_testX  = df_X_trimmed[df_X['Opportunity_ID'].isin(test_OPIDs)]

df_trainy = df_y[df_X['Opportunity_ID'].isin(train_OPIDs)]
df_valy   = df_y[df_X['Opportunity_ID'].isin(val_OPIDs)]
df_testy  = df_y[df_X['Opportunity_ID'].isin(test_OPIDs)]
```

```python
import sklearn.svm
import sklearn.ensemble
import sklearn.linear_model

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

for model in [sklearn.svm.SVC(probability=True), sklearn.ensemble.RandomForestClassifier(), sklearn.linear_model.LogisticRegression()]:
  model = model.fit(df_trainX, df_trainy)
  for x, y in zip([df_trainX, df_valX], [df_trainy, df_valy]):
    print(model)
    predictions = model.predict(x)

    plt.title(f"Confusion matrix")
    sn.heatmap(sklearn.metrics.confusion_matrix(y, predictions), annot=True)
    plt.xlabel(f"Predicted")
    plt.ylabel(f"Actual")
    plt.show()

    print(model.predict_proba(x))
```
